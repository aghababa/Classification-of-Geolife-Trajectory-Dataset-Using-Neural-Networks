```python
float = np.vectorize(float)

def read_file(file_name):
    data = []
    with open(file_name, "r") as f:
        for line in f:
            item = line.strip().split(",")
            data.append(float(item))
    return data
```

# Classifiers

```python
clf0 = KNeighborsClassifier(n_neighbors=5) 
clf1 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'linear', max_iter = 200000))
clf2 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'rbf', gamma= 'auto', max_iter = 200000))
clf3 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'rbf', gamma= 1, max_iter = 200000))
clf4 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'rbf', gamma= 10, max_iter = 200000))
clf5 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'rbf', gamma= 100, max_iter = 200000))
clf6 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'rbf', gamma= 1000, max_iter = 200000))
clf7 = make_pipeline(StandardScaler(), svm.SVC(kernel = 'poly', max_iter = 400000))
clf8 = DecisionTreeClassifier()
clf9 = RandomForestClassifier(n_estimators=50)
clf10 = AdaBoostClassifier(n_estimators= 50,learning_rate=1)

clf = [clf0,clf1,clf2,clf3,clf4,clf5,clf6,clf7,clf8,clf9,clf10]
```

## Training

```python
Start_time = time.time()

mean_train_error = [0] * len(clf)
mean_test_error = [0] * len(clf)
var_test_error = [0] * len(clf)

num_trials = 10

for k in range(len(clf)):
    Start_time_1 = time.time()
    for s in range(num_trials):

        X_train_labeled = [0] * r
        X_test_labeled = [0] * r
        error_train_list_all = []
        error_test_list_all = []

        Start_time_2 = time.time()

        for i in range(r-1): 

            X_train_labeled[i] = np.concatenate((X_train[i], np.array([[-1]]* len(X_train[i]))), axis =1)
            X_test_labeled[i] = np.concatenate((X_test[i], np.array([[-1]]* len(X_test[i]))), axis =1)
            j = i+1
            X_train_labeled[j] = np.concatenate((X_train[j], np.array([[1]]* len(X_train[j]))), axis =1)
            X_test_labeled[j] = np.concatenate((X_test[j], np.array([[1]]* len(X_test[j]))), axis =1)

            data_fin_train = np.insert(X_train_labeled[i], len(X_train_labeled[i]), X_train_labeled[j], axis = 0)
            Z = list(data_fin_train)
            random.shuffle(Z)
            data_final_train = np.array(Z)

            data_final_test = np.insert(X_test[i], len(X_test[i]), X_test[j], axis = 0)
            data_final_test_labels = np.insert(X_test_labeled[i][:,-1], len(X_test[i]), X_test_labeled[j][:,-1], axis = 0)

            model = clf[k]

            #Train the model using the training sets
            model.fit(data_final_train[:,:-1], data_final_train[:,-1])

            #Predict the response for test dataset
            y_pred = model.predict(data_final_test)
            ee = 1 - metrics.accuracy_score(data_final_test_labels, y_pred)
            error_test_list_all.append(ee)

            x_pred = model.predict(data_final_train[:,:-1])
            ff = 1 - metrics.accuracy_score(data_final_train[:,-1], x_pred)
            error_train_list_all.append(ff)

    mean_train_error[k] += np.mean(error_train_list_all)
    mean_test_error[k] += np.mean(error_test_list_all)
    var_test_error[k] += np.var(error_test_list_all)

    print("train error for clf["+str(k)+"] = ", np.mean(mean_train_error[k]))
    print("test error for clf["+str(k)+"] = ", np.mean(mean_test_error[k]))
    print("std of test error for clf["+str(k)+"] = ", np.mean(var_test_error[k]))

    print('time for step', s, time.time() - Start_time_2)

print('time for step', s, time.time() - Start_time_1)

print('total time =', time.time() - Start_time)
```

### dim =20, sigma = 0.3

```python
print('|Q|=20,', 'sigma = 0.3,', 'Number of Trials = ', num_trials)

Dic1 = {}

models = ["KNN (n_neighbors=5)","Linear SVM", "Gaussian kernel SVM, gamma=auto", 
          "Gaussian kernel SVM, gamma=1", "Gaussian kernel SVM, gamma=10", 
          "Gaussian kernel SVM, gamma=100", "Gaussian kernel SVM, gamma=1000", 
          "Poly SVM, deg= auto", "Decision Tree"]

for k in range(len(models)): 
    Dic1[k+1] = [models[k], np.round(np.mean(mean_train_error[k])/num_trials, decimals = 4), 
                np.round(np.mean(mean_test_error[k])/num_trials, decimals = 4),
                np.round(np.mean(var_test_error[k])/num_trials, decimals = 4)]
    
df1 = pd.DataFrame.from_dict(Dic1, orient='index', columns=['Classifier','Train Error', 
                                                        'Test Error', 'Variance Error'])
df1
```
